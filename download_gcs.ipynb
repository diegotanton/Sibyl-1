## **Download from GCS**

```python
from google.cloud import storage
from google.oauth2 import service_account
import os
```

```python
## Check GCP Credentials

credentials_path = os.getenv('GOOGLE_APPLICATION_CREDENTIALS')
if credentials_path is None:
    print("Credentials path not set.")
else:
    print(f"Credentials path: {credentials_path}")
    # Verify if the file exists
    if os.path.exists(credentials_path):
        print("Credential file found.")
        try:
            credentials = service_account.Credentials.from_service_account_file(
                credentials_path
            )
            print("Credentials are valid.")
        except Exception as e:
            print(f"Error loading credentials: {e}")
    else:
        print("Credential file does not exist.")
```

```python
## Download All Files from GCS Bucket

def download_all_files(bucket_name, destination_folder):
    storage_client = storage.Client()
    bucket = storage_client.get_bucket(bucket_name)

    blobs = bucket.list_blobs()
    for blob in blobs:
        file_path = os.path.join(destination_folder, blob.name)
        os.makedirs(os.path.dirname(file_path), exist_ok=True)
        blob.download_to_filename(file_path)
        print(f"Downloaded {blob.name} to {file_path}")

bucket_name = "dt-newsfeed-42u842shu7i21ao1pt8m"
destination_folder = "data/industrials_073124/"
download_all_files(bucket_name, destination_folder)
```
